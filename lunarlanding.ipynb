```python
# this block imports the algo we will be using, A2C. A2C uses a combination of policy based and value based methods,to put it simply, the agent will do some action and once it finishes, a "outside source" will let you know if that action was a good or bad aciton hence the actor and critic name
# A2C had to functions: a policy to control the agent and value function to tell how good the action was, which updates the policy of the agent
# a wrapper that goes around the GYM env and vectorizes it, this will help us train the agent n environment per step
# tool to help us understand how the model is doing

from stable_baselines3 import A2C
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
env = gym.make("LunarLander-v2")
```

```python
#this block of code, run the lunar landing game 10 times, there are varibales like state, done, and score to evaluate the environmnet,
#the while loops runs the env and returns the result of the simulation

episodes = 10
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0

    while not done:
        env.render()
        action = env.action_space.sample()
        obs, reward, done, truncated, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
```

```python
#we wrap the env in a dummyvecenv
#making a instance of A2C
#mlppolicy is a multilayer perceptron, a feedforward nn, consisting of fully connected neruons with nonlinear activation function
env = gym.make("LunarLander-v2")
env = DummyVecEnv([lambda : env])
model = A2C('MlpPolicy', env, verbose = 1)
```

```python
model.learn(total_timesteps=100000)
```

```python
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=100)
print(f"mean_reward:{mean_reward:.2f} +/- {std_reward:.2f}")
env.close
```
